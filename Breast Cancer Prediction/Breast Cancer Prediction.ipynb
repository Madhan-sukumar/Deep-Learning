```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

## Data Collection:

```python
# loading the data from sklearn
dataset = sklearn.datasets.load_breast_cancer()
```

```python
# loading the data to a dataframe

df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

# Data Splitting

```python
X = df.drop(columns='target',axis=1)
Y = df['target']
```

```python
print(X.shape,Y.shape)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X_train.shape,X_test.shape)
```

```python
print(Y_train.shape,Y_test.shape)
```

# Standardize the Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

#should fit only for training data
X_train_std = scaler.fit_transform(X_train)

#test data shouldn't fit it undergoes only transform
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

## Building the Neural Network

```python
# importing tensorflow and keras

import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# creating the layers

model= keras.Sequential([
                         ## first layer - converting into 1d array for 30 features
                         keras.layers.Flatten(input_shape=(30,)), 
    
                         ## Hidden layer with 20 neurons
                         ## relu function is used to eliminate negative's
                         keras.layers.Dense( 20, activation ='relu'),
    
                         ## Output layer with 2 ouput neurons beacuse dataset contains only 2 target classes
                         ## sigmoid function used for binay classification (logistic regression) 
                         keras.layers.Dense(2, activation ='sigmoid')   
])
```

```python
# compiling the neural network

model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',  ## loss function
             metrics=['accuracy'])
```

# Training the NN:

```python
# training the neural network

history = model.fit(X_train_std,Y_train, validation_split=0.1, epochs =10)
```

# visualizing the accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['training data accuracy','validation data accuracy'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['training data loss','validation data loss'], loc='upper right')
```

# Testing the NN:

```python
#model predicts and compares the predicted value with original data by itself
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
y_pred = model.predict(X_test_std)
```

```python
print(y_pred)
```

```python
# now converting the prediction probability to class labels

y_pred_labels = [np.argmax(i) for i in y_pred]
print(y_pred_labels)
```
